```python
# import all that require
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Prediction with regression linear
df = pd.read_csv('ds_salaries.csv')
df.head()
```

```python
#filter the data for job_title do Data Science
Data_Scientist = df[df['job_title'] == 'Data Scientist']
print(Data_Scientist)
```

```python
work_year = Data_Scientist['work_year']
salary_in_usd = Data_Scientist['salary_in_usd']
```

```python
x = np.array(salary_in_usd).reshape(-1, 1)
y = np.array(work_year)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=20, random_state=42)
```

```python
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
y_prediction = regression.predict(x_train)
mse = mean_squared_error(y_train, y_prediction)
print("Mean Squared Error:", mse)

prediction = regression.predict(x_test)
```

```python
# melihat perbandingan harga antara Data possible salary data science
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, regression.predict(x_train), color = 'red')
plt.title("Data Science possible salary")
plt.xlabel('year')
plt.ylabel('Salary in USD')
plt.show()

plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, regression.predict(x_train), color = 'red')
plt.title("Data Science possible salary")
plt.xlabel('year')
plt.ylabel('Salary in USD')
plt.show()
```
